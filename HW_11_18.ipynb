2) Describe what is version control in your own words and how it could help you in your project
   
**It's a tracking system that records any changes made to files over time and by which user. It helps keep track of any changes I make and if I run into any issues writing the code, I can go back to an earlier idea or script that I wrote that worked much better.**

3) Explain how in a group project (which we are not doing) github could help manage collaboration.
   
 **It would help each member to access the project without issues or authorization or overwriting eachother's code. It would also help keep track of changes by each member on any of the software, documentation, pipelines, or projects. It would also help that each member doesn't need to proceed one step at-a-time too.**
